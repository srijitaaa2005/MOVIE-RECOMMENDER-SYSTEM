```python
#genres 
#id
#keywords
#title
#overview
#cast 
#crew

#keeping these columns and removing the rest
movies = movies[['movie_id','title','genres','overview', 'keywords', 'cast', 'crew']]
movies.head()
```

```python
movies.isnull().sum() #checking for missing data
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.iloc[0].genres #checking genre of the first movie
```

```python
# extracting just the genre name of the movies in a list
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj): #string becomes a list
        L.append(i['name'])
    return L 
```

```python
movies['genres']=movies['genres'].apply(convert)
movies.head()
```

```python
movies.iloc[0].keywords
```

```python
#extracting on the keyword names of the movies
movies['keywords']=movies['keywords'].apply(convert)
movies.head()
```

```python
movies.iloc[0].cast
```

```python
#extracting the first 3 actors from cast
def actors(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj): #string becomes a list
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else: break
    return L 
```

```python
movies['cast']=movies['cast'].apply(actors)
movies.head()
```

```python
movies.iloc[0].crew
```

```python
#extracting director names only (job=director)
def fetch_director(obj):
    L=[]
    
    for i in ast.literal_eval(obj): #string becomes a list
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L 
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies.head()
```

```python
#a = s.split()
movies['overview']=movies['overview'].apply(lambda x:x.split()) #converting overview column from string to list
movies.head()
```

```python
#removing spaces from the columns to remove confusion
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['overview']=movies['overview'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])

movies.head()
```

```python
#making a "tags" column
movies['tags']=movies['genres']+movies['overview']+movies['keywords']+movies['cast']+movies['crew']
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
new_df.head()
```

```python
movies.iloc[0].tags
```

```python
#converting the tags column from list to string
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x)) 
#for each row, takes the list of words and joins them into one single string, separated by spaces
#still a list but now contains a combined string

new_df.head()
```

```python
new_df.iloc[0].tags
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower()) #converting tags columns into lowercase
new_df.head()
```

```python
#removing similar words like actor,actors
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df.iloc[0].tags
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=4500,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
vectors.shape
```

Calculating cosine distance

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
similarity[1]
```

```python
similarity[2]
```

```python
#in order to maintain index position
list(enumerate(similarity[0]))
```

```python
#reverse sorting on the basis of cosine distances
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1]) [1:6]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0] #fectching movie index
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1]) [1:6]

    if movie not in new_df['title'].values:
        print("Movie not found in database.")
        return

    for i in movies_list:
        print(new_df.iloc[i[0]].title) #generate the movie title from the index
```

```python
#recommend('Harry Potter')
recommend("Avatar")
```
